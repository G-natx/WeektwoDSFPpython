```python
#10. Write a function that counts vowels in a string.
def vowel_count(text):
    count = 0
    for letter in text:
        if letter in 'aeiou':
            count += 1
    return count
```

```python
#11. Write a function that takes a list and returns a new list with unique values.
def unique_values(main_list):
  """
This function is designed to take a list and return a new list containing only the unique values.
  """
  # Convert the list to a set to automatically remove duplicates
  unique_set = set(main_list)

  # Convert the set back to a list to return
  unique_list = list(unique_set)

  return unique_list

```

```python
unique_values([5, 4, 3, 2, 5, 3, 4])
```

```python
#24. Write a function that calculates the power of a number.
def calc_power(base, exponent):
    result = 1
    for _ in range(abs(exponent)):
        result *= base
    return result if exponent >= 0 else 1 / result
print(calc_power(2, 4))
```

```python
#27. Write a function that checks if a number is prime.
def prime_num(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

prime_num(20)
```

```python
#29. Write a function that accepts a list of prices and applies a default discount.

def apply_default_discount(prices, discount_rate=0.10):
    """
    This function returns a list of prices after applying a discount on them.
    
    """
    discounted = []

    for price in prices:
        discounted_price = price - (price * discount_rate)
        discounted.append(discounted_price)

    return discounted

```
